```python
import librosa
from obspy import read
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from obspy import UTCDateTime, read, Trace, Stream
import glob
import obspy
import obspy.signal
import random
```

```python
#STFT of data, stack (for using it as 1D time series)

directory= '/Users/zali/study/data/iceland/Reykjanes/Reykjanes/NUPH/HHE/stackSTFT/*.mseed'
folder= sorted (glob.glob (directory))

hop_length=225
n_fft=win_length=512


stftstack= []
for file in folder:
    st= read (file)
    for l in range (len (st)):
        st[l].detrend('demean')
        st[l].filter('bandpass', freqmin=1.0, freqmax=5.0, zerophase=True)
        st[l].resample(8.0)
        idx=len(st[l])- 57600
        if idx>0:
            x=st[l].data
        
            for i in range (0, 100): # 100 random STFT for each day
                j= random.randint (0,idx) #random start time from the bigining to 2 hours before the end of the stream
                y=x[j:j+57600].astype(float)
                s= librosa.stft(y, n_fft= n_fft, hop_length=hop_length, win_length=win_length)
                s2 = np.zeros((256, 256))
                s2 = s[1:, :256]
                with open ('stack2hour-'+file[70:-6]+'-'+ str (i+1) + '-' + str (l) + '.txt','w') as h:
                    out = (np.abs(s2)).sum(axis=0)
                    h.write("\n".join(map(lambda x: '%f' % x, out)))
                    stftstack.append (out)
                    np.save('stft', out)
                fig= plt.figure(figsize=(15, 5))
                librosa.display.specshow(librosa.power_to_db(s, ref=np.max),y_axis='linear', sr=8)
                plt.title(str (st[l].stats.starttime)+ '-'+ str (i+1) + 'th' + '-2hours')
                plt.ylabel('Frequency [Hz]', fontsize= 16)
                plt.xlabel('Time [hour]', fontsize= 16)
                plt.yticks ([0,1,2,3,4], fontsize= 16)
                plt.xticks (fontsize= 16)
                labelsx = [0,0.5,1,1.5,2]
                plt.xticks(np.arange(0,257, 64), labelsx, fontsize= 16)
                plt.ylim(1,4)
                fig.savefig ( 'stack2hour-'+file[70:-6]+'-'+ str (i+1) + '-' + str (l) + '.png', dpi= 100)
                #plt.show ()
                plt.close ()
                stft= np.stack (stftstack)
    np.save('randstackSTFT_2may-24june.npy', stft)                    
            
```

```python
# apply HPS for removing the EQs: one horizontal median filter

hop_length=8
win_length=n_fft=32
new= []

x = np.load('randstackSTFT_2may-24june.npy')

for i in range (0,5800):
    
    # Compute the spectrogram amplitude and phase
    S_full, phase = librosa.magphase(librosa.stft(x[i], n_fft= n_fft, hop_length=hop_length, win_length=win_length))


    D_harmonic, D_percussive = librosa.decompose.hpss(S_full,kernel_size= 256,  margin=[1,1])
    
    
    h=D_harmonic*phase
    L = x[i].shape[0]
    harm = librosa.istft(h, hop_length=hop_length, win_length=win_length, window='hann', length=L)
    new.append (harm)
    median= np.stack (new)
np.save('med-randstackSTFT_2may-24june.npy', median) 
```

```python
# individual normalization

norm= []
x = np.load('med-randstackSTFT_2may-24june.npy')
for i in range (0,5800):
    
    normcal= (x[i] - np.min(x[i])) / (np.max(x[i]) - np.min(x[i]))
    norm.append (normcal)
    out= np.stack (norm)
np.save('innorm-med-randstackSTFT_2may-24june.npy', out)  
```

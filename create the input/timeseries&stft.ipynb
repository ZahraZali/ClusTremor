```python
import librosa
from obspy import read
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from obspy import UTCDateTime, read, Trace, Stream
import glob
import math
```

```python
# create events equal to cycle duration = duration + repose time


t1=UTCDateTime('2021, 5, 2, 00, 00, 00')
t2=UTCDateTime('2021, 6, 13, 21, 11, 00')
#st = read('RR*', starttime=t1, endtime=t2)
#st.merge(method=1, fill_value='interpolate')

markerfile = "2022_010_Eibl-et-al_Data-modified(markers in gaps are deleted).txt"


directory= '/Users/zali/study/data/iceland/Reykjanes/Reykjanes/NUPH/HHE/12july/events2.5-13.6/*.mseed'
folder= sorted (glob.glob (directory))
for file in folder:
    st = read('RR*', starttime=t1, endtime=t2)
    st.merge(method=1, fill_value='interpolate')   

    with open(markerfile, 'r+') as f:
        lines = f.readlines()[10:]
        for i in range(0, len(lines)-1):
            if i % 2 == 0:
                st2 = st.copy()
                line1 = lines[i]
                t=line1[0:10]+'T'+line1[11:25]
                s= UTCDateTime(t)-10  #startt time 10 seconds before each event
                line2 = lines[i+1]  # end time of the event
                e=line2[0:10]+'T'+line2[11:25]
                st3=st2.trim (s , endtime=UTCDateTime(e))
                st3.write( str (int(math.ceil((i+1)/2))) + '_' + str (t)[:-5] + 'mseed', format='MSEED')
```

```python
# save events duration in a text file

directory= '/Users/zali/study/data/iceland/Reykjanes/Reykjanes/NUPH/HHE/12july/events2.5-13.6/*.mseed'
folder= sorted (glob.glob (directory))
with open('cycleduration+10s', 'w+') as f:
    for file in folder:
        st= read (file)
        f.write(str(st[0].stats.npts)+ "  " + str(st[0].stats.starttime) + "  " + str(st[0].stats.endtime) + "\n")
f.close ()
```

```python
# remove the 20 longest events and zero padding others

directory= '/Users/zali/study/data/iceland/Reykjanes/Reykjanes/NUPH/HHE/12july/events2.5-13.6/*.mseed'
folder= sorted (glob.glob (directory))
for file in folder:
    st= read (file)
    duration= st[0].stats.npts
    if duration< 167614:
        st[0].data= np.pad (st[0].data, (0, 165498), 'edge')
        stnew2= st.trim (starttime=st[0].stats.starttime, endtime=st[0].stats.starttime+827.49)
        stnew2.write ( '14min-' + file[82:], format='MSEED')
```

```python
# save all cycles as np file

cycles= []
directory= '/Users/zali/study/data/iceland/Reykjanes/Reykjanes/NUPH/HHE/12july/cycles2.5-13.6/cycles2.5-13.6zeropadded/*.mseed'
folder= sorted (glob.glob (directory))
for file in folder:
    st= read (file)
    st[0].resample(10.0)
    st[0].detrend('demean')
    st[0].filter('bandpass', freqmin=1.0, freqmax=4, zerophase=True)
    x=st[0].data
    cycles.append (x[:-45])
    np.save('cycles', x[:-45])
    
data= np.stack (cycles)
np.save('21mincycles-ds10hz-filt1-4_2.5-13.6.npy', data)
```

```python
# STFT of the cycles

directory= '/Users/zali/study/data/iceland/Reykjanes/Reykjanes/NUPH/HHE/12july/events2.5-13.6/events2.5-13.6zeropadded/*.mseed'
folder= sorted (glob.glob (directory))

hop_length=1300
win_length=n_fft=4096


stft_list= []
for file in folder:
    st= read (file)
    x=st[0].data
    y=x.astype(float)
    s= librosa.stft(y, n_fft= n_fft, hop_length=hop_length, win_length=win_length)
    s2 = s[19:83,:]  # frequency band [1-4] Hz
    stft_list.append (s2)
    np.save('stft', s2)
    
stft= np.stack (stft_list)
np.save('events-stft14min_2.5-13.6.npy', stft)    
```

```python
# individual normalization

norm= []
x = np.load('events-stft14min_2.5-13.6.npy')
for i in range (0,6993):
    
    normcal= (x[i] - np.mean(x[i])) / np.std(x[i])
    norm.append (normcal)
    out= np.stack (norm)
np.save('innorm-events-stft14min_2.5-13.6.npy', out)  
```
